```python
col = [ 'pos_pct', 'neg_pct', 'neutral_pct']

# Create the df_text_score DataFrame
df_text_score = pd.DataFrame(columns = col)
```

```python
import warnings
warnings.filterwarnings("ignore")
from textblob import TextBlob

# Create a list of your 10 DataFrames (df1, df2, ..., df10)
dataframes = [df1 ,df2, df3, df4, df5, df6, df7, df8, df9, df10]

# Initialize counters for positive, negative, and neutral text
positive_count = 0
negative_count = 0
neutral_count = 0

# Iterate through the DataFrames and calculate sentiment percentages
for df in dataframes:
    for index, row in df.iterrows():
        text = row['text']
        blob = TextBlob(text)
        
        # Perform sentiment analysis
        sentiment_score = blob.sentiment.polarity
        
        if sentiment_score > 0:
            positive_count += 1
        elif sentiment_score < 0:
            negative_count += 1
        else:
            neutral_count += 1
    # Calculate the total count of texts across all DataFrames
    total_texts = positive_count + negative_count + neutral_count
    # Calculate the percentage of positive, negative, and neutral texts
    positive_percentage = (positive_count / total_texts) * 100
    negative_percentage = (negative_count / total_texts) * 100
    neutral_percentage = (neutral_count / total_texts) * 100
    # Inserting into df_text_score
    new_row = {'pos_pct': positive_percentage, 'neg_pct': negative_percentage, 'neutral_pct': neutral_percentage}

    # Append the new row to the DataFrame
    df_text_score = df_text_score.append(new_row, ignore_index=True)
    positive_count = negative_count = neutral_count =0
    

```

## Decision Making - Should we select this candidate
#### Based upon his communication skills we can select this candidate. Because this candidate looks neutral and positive most of the time with his words. Also he shows enthusiasm during his achievements, awards and passion. His constant speech speed, word selections showing him confident (good confident score for 50% of his speech time also not very less confident score for 25% time), positive, less negative words about himself(50 percentile score = 0.08) and less hesitant in delivering his speech(50percentile score = 0.4).

```python
def calculate_candidate_selection_score(pos_pct, neg_pct, neu_pct, hesitancy_mean, hesitancy_std, enthusiasm_mean, enthusiasm_std, confidence_mean, confidence_std, conciseness_mean, conciseness_std, speech_speed_mean, speech_speed_std):
    # Define weights for each parameter
    weights = {
        'pos_pct_weight': 3,
        'neg_pct_weight': -3,
        'neu_pct_weight': 1,
        'hesitancy_weight': -2,
        'enthusiasm_weight': 4,  # Adjusted weight for enthusiasm
        'confidence_weight': 3,
        'conciseness_weight': 3,
        'speech_speed_weight': 1.5,  # Adjusted weight for speech_speed
    }

    # Calculate Z-scores for hesitancy, confidence, conciseness
    hesitancy_zscore = (hesitancy_mean - 0) / hesitancy_std
    confidence_zscore = (confidence_mean ) / confidence_std
    conciseness_zscore = (conciseness_mean ) / conciseness_std

    # Calculate absolute Z-scores for enthusiasm and speech_speed
    enthusiasm_zscore = (enthusiasm_mean ) / enthusiasm_std
    speech_speed_zscore = abs(speech_speed_mean-3) / speech_speed_std
    # Assuming that 3 is best speed to communicate. 

    # Calculate weighted scores for each parameter
    pos_pct_score = pos_pct * weights['pos_pct_weight']/10 # (Maximum points = 30)
    neg_pct_score = neg_pct * weights['neg_pct_weight']/10 # (Minimum points = -20)
    neu_pct_score = neu_pct * weights['neu_pct_weight']/10 # (Maximum points = 10)
    hesitancy_score = hesitancy_zscore * weights['hesitancy_weight']
    enthusiasm_score = enthusiasm_zscore * weights['enthusiasm_weight']
    confidence_score = confidence_zscore * weights['confidence_weight']
    conciseness_score = conciseness_zscore * weights['conciseness_weight']
    speech_speed_score = speech_speed_zscore * weights['speech_speed_weight']

    # Calculate overall candidate score by summing the weighted scores
    overall_score = (pos_pct_score + neg_pct_score + neu_pct_score +
                     hesitancy_score + enthusiasm_score + confidence_score +
                     conciseness_score + speech_speed_score)

    # Define a threshold for candidate selection (you can adjust this as needed)
    selection_threshold = 34  # Assumed to be 30 and can be adjusted
    if overall_score > selection_threshold : 
        print('Selected on the basis of Transcript analysis')
    else:
        print('Not Selected on the basis of Transcript analysis')
    print('Transcript score is ', overall_score)
              
        
```
